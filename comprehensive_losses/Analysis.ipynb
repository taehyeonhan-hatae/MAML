```python
#!pip install matplotlib
#!pip install pandas

import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_comprehensive_loss = pd.read_csv('maml_comprehensive_losses.csv')
df_comprehensive_loss_train = df_comprehensive_loss[df_comprehensive_loss['phase'] == 'train']
df_comprehensive_loss_val = df_comprehensive_loss[df_comprehensive_loss['phase'] == 'val']
```

```python
df_comprehensive_loss_train
```

```python
df_comprehensive_loss_train[['support_loss_0']]
```

```python
df_comprehensive_loss_train[['support_loss_0']].plot(
    xlabel='x',
    ylabel='support_loss_0',
    title='support_loss_0'
)

plt.show()
```

```python
df_comprehensive_loss_train[['target_accuracy_4']].plot(
    xlabel='x',
    ylabel='target_accuracy_4',
    title='target_accuracy_4'
)

plt.show()
```

```python
df_comprehensive_loss_train[['support_accuracy_4', 'target_accuracy_4']].plot(
    xlabel='x',
    ylabel='acc',
    title='(Inner-loop training) Support VS target'
)

plt.show()
```

```python
df_comprehensive_loss_val
```

```python

df_comprehensive_loss_val[['support_accuracy_4', 'target_accuracy_4']].plot(
    xlabel='x',
    ylabel='acc',
    title='(Inner-loop validation) Support VS target'
)

plt.show()
```

```python
df_summary_statistics = pd.read_csv('maml_summary_statistics.csv')

df_summary_statistics[['epoch','train_loss_mean', 'val_loss_mean']].plot(
    x = 'epoch',
    xlabel='epoch',
    ylabel='loss',
    title='(outer-loop) train_loss_mean VS val_loss_mean'
)

plt.show()
```

```python
df_summary_statistics[['epoch','train_accuracy_mean', 'val_accuracy_mean']].plot(
    x = 'epoch',
    xlabel='epoch',
    ylabel='acc',
    title='(outer-loop) train_accuracy_mean VS val_accuracy_mean'
)

plt.show()
```
